* Full of 30,000 records

##### Attributes 
* Most important ones: 
    * Blood pressure (Decimal)
    * BMI (Body Mass Index: Decimal)
    * Length of Stay (hospital length of stay (in days))
    * Family history (binary: 0 (no) OR 1 (yes))
        * Does not go into much detail
    * Cholesterol (the level)
    * Oxygen Saturation 
    * Age, Gender
* OTHERS: (might be noise)
    * Glucose, hbA1c, Smoking, Alchohol, Stress Level, Sleeep Hours, Triglycerides, Diet Score, Physical Activity 
    * two defininitely noisy columns: "random_notes" (a lot of values missing), "noise_col"

* What will we be classifying? 
    * Medical Condition: 
        * EX: Diabetes, Hypertension, Asthma, Obesity 
    * Other options: Blood Pressure, LengthOfStay, Diet Quality Score
